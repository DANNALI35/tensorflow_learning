在实际应用中，在某个计算机视觉任务表现良好的神经网络在其他视觉任务上通常也适用。本节简要了解3个经典cnn神经网络：**LeNet-5,AlexNet，VGG**以及近期功能较为强大的**ResNet、InceptionV1~V4**，为后续图片识别工作获得一些灵感。

## LeNet-5  
LeNet-5于1998年发表，论文地址在http://yann.lecun.com/exdb/publis/pdf/lecun-01a.pdf ，应用于手写数字识别，其网络结构如下：

 LeNet-5主要特点：
* 卷积时padding=0，属于有效卷积
* 池化窗口(2,2)，每次将图像变量的高度和宽度减小一半(**至今常用**)
* 网络越深，图像变量的高度和宽度变小，通道数增加(**至今常用**)
* 网络结构为‘卷积层1+池化+卷积层2+池化...+全连接层1+全连接层2+输出层’(**至今常用**)
* 卷积层、池化层激活函数分别是tanh、sigmoid
* 参数约60,000
* 池化方法是平均池化
* 多层神经网络（MLP）作为最后的分类器


## AlexNet
AlexNet于2012年发表，论文地址为https://papers.nips.cc/paper/4824-imagenet-classification-with-deep-convolutional-neural-networks.pdf ，应用于ImageNet 2010年LSVRC数据比赛中120万图片的1000个类别分类。该论文在计算机视觉和其他领域造成了巨大影响，人们从此认可深度学习在计算机视觉中占有一席之地。AlexNet结构如下：

**AlexNet主要特点：**
* 5个卷积层，第一次卷积padding=0，后四次卷积是same卷积
* 参数约6千万
* 卷积层激活函数为Relu(**至今常用**)
* 输出层激活函数为Softmax(**至今常用**)
* 池化方法是最大池化(**至今常用**)

## VGG-16
VGG-16于2015年发表，论文地址为https://arxiv.org/pdf/1409.1556.pdf ， 应用于imagenet ILSVRC2014比赛中的图片物体识别，VGG-16网络在该比赛中为第一名算法，其网络结构见下图。 

**VGG-16主要特点：**
* 16个层网络的参数是训练得来的，
* 5个卷积单元，13个卷积层，随着网络层数增加，滤波器数量以*2的速度增加，达到512后保持不变
* 约1380万参数  
**注**：VGG-19比VGG-16多3个卷积层，两者效果差异不大

## ResNet

ResNet（残差网络）发表于2015年，论文地址为https://arxiv.org/pdf/1512.03385.pdf ，该网络在2015年imagenet各大赛事上都拔得头筹。以下对论文中ResNet思想和实验作简要介绍。

在现实深度学习问题中，更深的模型常产生比它更浅的模型更高的错误率。而这个“退化”问题产生的原因归结于优化难题，当模型变复杂时，SGD的优化变得更加困难，导致了模型达不到好的学习效果。

为解决以上为题，Residual Networks中引入了residual block的思想，即通过shortcut将每个block的输入（下图中的$a^{[l]}$)和输出下图中的（$a^{[l+1]}$)进行一个element-wise的加叠，这个block中的residual为$a^{[l]}$，这个简单的加法并不会给网络增加额外的参数和计算量，同时却可以大大增加模型的训练速度、提高训练效果，并且当模型的层数加深时，这个简单的结构能够很好的解决退化问题。

VGG-19、plain网络、ResNet网络结构对比  

**34层ResNet特点**
* 34层的参数是训练的来的
* 卷积时，每两个卷积层加入一个residual block
* 网络越深，图像变量的高度和宽度变小，通道数增加
* 卷积激活函数是relu
* 参数量36万

下图为论文中的实验结果，在plain上观测到明显的退化现象，而且ResNet上不仅没有退化，34层网络的效果反而比18层的更好，而且不仅如此，ResNet的收敛速度比plain的要快得多。

论文中试验了18、34、50、101、152层的ResNet，这些网络而且不仅没有出现退化问题，错误率也大大降低，同时计算复杂度也保持在很低的程度。下图为实验各层数ResNet的结构以及Residual block设置情况。

## Inception Network

### Inception Network结构介绍

Inception Network的主要思想：**在网络同一层同时进行不同大小滤波器卷积和池化**，从而避免不同大小滤波器的选择和卷积层、池化层的排列。下图为单个Inception模块结构示意图。

Inception Network存在问题：计算次数过多

解决问题：使用(1,1)卷积减少计算次数

## Inception-V1(GoogleNet)

inception network于2014年在论文Going deeper with convolutions中提出，论文地址为https://arxiv.org/pdf/1409.4842.pdf 。下图为论文中inception network 结构。

**图注释**
* 蓝色框表示卷积过程
* 红色框表示池化过程
* 黄色框表示全连接层
* 椭圆黑色框表示softmax输出层

**Inception network特点**
* 由多个inception模块组合
* 隐藏层中间分支是提前用全连接层和softmax作预测，起到正则化的作用

目前inception network已经出现了一些新版本比如inception-V2、inception-V3、inception-V4，这些新版本都是通过一定的方式将单个inception模块组合到一起。另外还有inception和ResNet的Residual block思想结合的网络，其表现通常更好。



## Inception-V2

Inception-v2单个模块结构如下图。在Inception-V1的（5，5）的卷积层用两个（3，3）卷积层代替，在减少参数同时增强了非线性表示能力。另外，Inception-v2提出了Batch Normalization（批规范化）。即在每次SGD时，通过mini-batch来对相应的activation做归一化操作，使得结果（输出信号各个维度）的均值为0，方差为1。Batch Normalization 是一个非常有效的正则化方法，可以让大型卷积网络的训练速度加快很多倍，同时收敛后的分类准确率也可以得到大幅提高。

## Inception-V3
inception-v3引入了 Factorization into small convolutions 的思想，将一个较大的二维卷积拆成两个较小的一维卷积，比如将3´3卷积拆成1´3卷积和3´1卷积，如下图所示。在减少参数同时增强了非线性表示能力，且加速运算并减轻了过拟合。另外。Inception V3 优化了 Inception Module 的结构。
inception-v3发表论文网址：https://arxiv.org/pdf/1512.00567  

## Inception-V4

inception-v4是inception-v3和ResNet的联合实现。inception-v4发表论文网址： https://arxiv.org/pdf/1602.07261.pdf

## 各种CNN参数量和准确率比较

图中的坐标轴我们可以看出横坐标是操作的复杂度，纵坐标是精度。模型设计一开始的时候模型权重越多模型越大，其精度越高，后来出现了resNet、GoogleNet、Inception等网络架构之后，在取得相同或者更高精度之下，其权重参数不断下降。值得注意的是，并不是意味着横坐标越往右，它的运算时间越大。在这里并没有对时间进行统计，而是对模型参数和网络的精度进行了纵横对比。
