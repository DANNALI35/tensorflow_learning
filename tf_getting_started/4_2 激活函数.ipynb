# 1、激活函数介绍
在隐藏层或输出层神经元中，输入的 inputs 通过加权求和后，还被作用了一个函数，这个函数就是激活函数 Activation Function。使用这个函数的意义在于给神经元引入非线性因素，使得神经网络可以任意逼近任何非线性函数，从而可以解决分类问题。

上图图左为sigmoid，图右为Tanh，Tanh函数可以看作是sigmoid函数的缩放和平移。Tanh函数是关于原点中心对称，取值范围为[-1,1]。

Tanh函数导数=1-Tanh(x)^2,因此Tanh函数可以减小梯度消失程度，但无法避免函数饱和时梯度变小。通常Tanh函数梯度效率略高于sigmoid函数。

Relu的神经元死亡问题：
当一个非常大的梯度流过一个 ReLU 神经元或learning rate设置很大，更新过参数之后，参数变为负数，如果X输入都是正值，这个神经元再也不会对任何数据有激活现象了，那么这个神经元的梯度就永远都会是 0。

## 2.4 PRelu

PRelu函数是Relu函数的改进，可以避免Relu函数神经元死亡问题。ReLU中当x<0时，函数值为0。而Leaky ReLU则是给出一个很小的负数梯度值Alpha，当Alpha=0.01时，是PRelu的特例leaky Relu。


# 激活函数选择
* 通常不单独使用sigmoid；
* 解决二分类问题时可以选用sigmoid函数作为输出层单元激活函数，Tanh函数作为其他层单元激活函数；
* Relu函数最为常用，但要注意其learning Rate不可以过大；
* 回归问题选用线性激活函数作为输出层单元激活函数，其他层单元激活函数通常选择非线性激活函数。
